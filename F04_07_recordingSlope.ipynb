```python
import os
files = []
folder = "train"
for file in os.listdir(folder):
    files.append(folder+"\\"+file)
```

```python
def slope(y):
    slop = []
    for i in range(1,len(y)):
        slop.append(y[i]-y[i-1])
        
    return slop
```

```python
y90, sr90 = librosa.load(files[90])

t90 = list(range(len(y90)))

plt.plot(t90,y90)
plt.show()
```

```python
slope90 = slope(y90)

ts90 = list(range(len(slope90)))

plt.plot(ts90,slope90)
plt.show()
```

```python
import wave, struct, math, random

def writeWav(file,sequence,sr):
    
    #sampleRate = 44100.0 # hertz
    #duration = 1.0 # seconds
    #frequency = 440.0 # hertz
    obj = wave.open(file,'w')
    obj.setnchannels(1) # mono
    obj.setsampwidth(2)
    obj.setframerate(22050)
    for i in range(len(sequence)):
        value = sequence[i]
        data = struct.pack("h", round(25000*value))
        obj.writeframes(data)
    obj.close()
    
    
```

```python
writeWav("y90sound.wav",y90,sr90)
```
